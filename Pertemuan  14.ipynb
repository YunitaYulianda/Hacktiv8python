# Unsupervised Learning

```python
# import dataset
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
# simpan datanya dlm new var
breast_data = breast.data
```

```python
breast_data.shape
```

```python
# simpan targetnya dlm new var
breast_label = breast.target
```

```python
breast_label.shape
```

```python
# import numpy agar bisa bentuk kembali hasil join breast_label dan breast_data  
# Dan akan dibuat DataFrame yg punya data dan label
import numpy as np
```

```python
# labels dibuat jdi dlm bentuk array 569  baris 1 kolom
labels = np.reshape(breast_label, (569,1))
labels
```

```python
# lalu digabung
final_breast_data = np.concatenate([breast_data, labels], axis=1)
```

```python
type(final_breast_data)
```

```python
final_breast_data.shape
```

```python
# dibuat dataframe
import pandas as pd
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
# print features
```

```python
features = breast.feature_names
```

```python
features
```

```python
features_labels = np.append(features,'label')
```

```python
features_labels
```

```python
# sematkan nama kolom ke dataframe 
breast_dataset.columns = features_labels
breast_dataset
```

```python
breast_dataset.head()
```

```python
# ubah nama label dri 0,1 jdi Benign dan Malignant
breast_dataset['label'].replace(0,'Benign', inplace=True)
breast_dataset['label'].replace(1,'Malignant', inplace=True)
```

```python
breast_dataset.head()
```

```python
breast_dataset.tail()
```

# Data Visualization using PCA

```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
features
```

```python
# ubah fitur jdi format tabel
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x, columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data=principalComponents_breast,
                                   columns=['principal component 1',
                                            'principal component 2'])
```

```python
principal_breast_Df.tail()
```

```python
print("Explained variation per principal component:{}".format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('PC - 1', fontsize = 20)
plt.ylabel('PC - 1', fontsize = 20)
plt.title("PCA of Breast Cancer", fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r','g']
for target, color in zip(targets, colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep,'principal component 1'],
                principal_breast_Df.loc[indicesToKeep,'principal component 2'],
                c = color, s = 50)
plt.legend(targets,prop={'size':15})
plt.show()
```
