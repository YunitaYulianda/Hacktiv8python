```python
# import library
from wordcloud import WordCloud, STOPWORDS
print('Wordcloud is installed and imported!')
```

```python
!pip install wget
```

```python
import wget
url = 'https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt'
filename = wget.download(url)
print(filename)
```

```python
# buka file
alice_novel = open('alice_novel.txt','r').read()
print('File downloaded and saved!')    
```

```python
stopwords = set(STOPWORDS)
```

```python
# tambahkan stopword
alice = WordCloud(background_color='white', max_words=2000, stopwords=stopwords)
# generate wordcloudnya
alice.generate(alice_novel)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# tampilkan wordcloudnya
plt.imshow(alice, interpolation = 'bilinear')
plt.axis('off')
plt.show()

# parameter interpolation berguna untuk kejelasan pixel gambar
# saat dilakukan zoom
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# said tidak informatif, maka better we delet it!
stopwords.add('said')

alice.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# wordcloud with masking
url= 'https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_mask.png'
filename = wget.download(url)
print(filename)
```

```python
alice_mask = np.array(Image.open('alice_mask.png'))
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# tambahkan stopword dan masknya
alice = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)
# generate wordcloudnya
alice.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)

plt.imshow(alice, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Choropleth Maps

```python
# masih use dataset tadi
df = pd.read_excel('Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
```

```python
# clean data
# hapus atr yang tidak penting
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df
```

```python
# ubah nama atr
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df
```

```python
# agar konsisten, maka ubah semua kolom jdi string
df.columns = list(map(str, df.columns))
```

```python
df['Total'] = df.sum(axis=1)
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df.head()
```

```python
# download geojson file
url = 'https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/world_countries.json'
filename = wget.download(url)
print(filename)
```

```python
world_geo = r'world_countries.json'

world_map = folium.Map(location=[0,0], zoom_start=2, tiles='Stamen Terrain')

world_map.choropleth(geo_data=world_geo,
                     data=df,
                     columns=['Country', 'Total'],
                     key_on = 'feature.properties.name',
                     fill_color = 'YlOrRd',
                     fill_opacity = 0.7,
                     line_opacity = 0.2,
                     legend_name='Immigration to Canada')

world_map
```
