# 1. Eksplor Data

```python
# dari dataset terdapat fitur seperti AREA, REG, DEV, Type, Coverage
# yang tidak berkonstribusi, jdi dihapus aja

df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df
```

```python
# melihat nilai dari kolom DevName
df['DevName'].unique()
```

```python
# rename kolom Odname, AreaName, dan RegName
df.rename(columns={'OdName': 'Country', 'AreaName':'Continent', 'RegName': 'Region'}, 
         inplace=True)
df.columns
```

```python
df
```

```python
# buat satu kolom baru yang menampung total imigran di suatu negara selama 33 tahun
df['Total']= df.sum(axis=1)
df
```

```python
# melihat info statistik
df.describe()
```

```python
# 4. Pie Chart
df_con = df.groupby('Continent', axis=0).sum()
print(type(df.groupby('Continent', axis=0)))

df_con.head()
```

```python
df_con['Total'].plot(kind='pie', figsize=(5,6), autopct='%1.1f%%', startangle=90, shadow=True)

plt.title('Immigration to Canada by Continent [1980-2013]')
plt.axis('equal')
plt.show()
```

```python
color_list= ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_con['Total'].plot(kind='pie', 
                     figsize=(5,6), 
                     autopct='%1.1f%%', 
                     startangle=90, 
                     shadow=True,
                     labels=None,
                     pctdistance=1.12,
                     colors=color_list,
                     explode=explode_list)

plt.title('Immigration to Canada by Continent [1980-2013]', y=1.12)
plt.axis('equal')
plt.legend(labels=df_con.index, loc='upper left')
plt.show()
```

```python
color_list= ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.05, 0.15]

df_con['2013'].plot(kind='pie', 
                     figsize=(15,6), 
                     autopct='%1.1f%%', 
                     startangle=90, 
                     shadow=True,
                     labels=None,
                     pctdistance=1.12,
                     colors=color_list,
                     explode=explode_list)

plt.title('Immigration to Canada by Continent Year 2013', y=1.12)
plt.axis('equal')
plt.legend(labels=df_con.index, loc='upper left')
plt.show()
```
